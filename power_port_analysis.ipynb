```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_half_power = pd.read_csv('output_default.csv')
df_full_power = pd.read_csv('full_power.csv')
df_ammar = pd.read_csv('ammar_suggestion.csv')
df_ammar_echo = pd.read_csv('ammar_suggestion1.csv')
```

```python
df_half_power[df_half_power['Metric'] == 'Throughput']
```

```python
# plotting the two dataframes distance vs throughput 

plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
sns.lineplot(data=df_half_power[df_half_power['Metric'] == 'Mean delay'], x='gnbUeDistance', y='Value', label='Half Power', color='blue')
sns.lineplot(data=df_full_power[df_full_power['Metric'] == 'Mean delay'], x='gnbUeDistance', y='Value', label='Full Power', color='orange')
plt.title('Distance vs delay')
plt.xlabel('Distance (m)')
plt.ylabel('delay (ms)')
plt.legend()
plt.grid(True)
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
sns.lineplot(data=df_half_power[df_half_power['Metric'] == 'Throughput'], x='gnbUeDistance', y='Value', label='Half Power', color='blue')
sns.lineplot(data=df_full_power[df_full_power['Metric'] == 'Throughput'], x='gnbUeDistance', y='Value', label='Full Power', color='orange')
sns.lineplot(data=df_ammar[df_ammar['Metric'] == 'Throughput'], x='gnbUeDistance', y='Value', label='ammar suggestion ', color='red')
sns.lineplot(data=df_ammar_echo[df_ammar_echo['Metric'] == 'Throughput'], x='gnbUeDistance', y='Value', label='ammar suggestion 50 percent  ', color='green')
plt.title('Distance vs througthput')
plt.xlabel('Distance (m)')
plt.ylabel('throughput (Mbps)')
plt.legend()
plt.grid(True)
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
sns.lineplot(data=df_half_power[df_half_power['Metric'] == 'Mean jitter'], x='gnbUeDistance', y='Value', label='Half Power', color='blue')
sns.lineplot(data=df_full_power[df_full_power['Metric'] == 'Mean jitter'], x='gnbUeDistance', y='Value', label='Full Power', color='orange')
plt.title('Distance vs delay')
plt.xlabel('Distance (m)')
plt.ylabel('delay (ms)')
plt.legend()
plt.grid(True)
plt.show()
```
