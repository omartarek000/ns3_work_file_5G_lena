```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_half_power = pd.read_csv('output_default.csv')
df_full_power = pd.read_csv('full_power.csv')
df_normal_run_allegation = pd.read_csv('normal_run_allegation.csv')
df_upnormal_run_allegation = pd.read_csv('upnormal_run_allegation.csv')
df_upnormal_run_allegation_16half = pd.read_csv('upnormal_run_allegation16half.csv')
df_upnormal_run_allegation_16 = pd.read_csv('upnormal_run_allegation16.csv')
df_normalizer_off_full_power = pd.read_csv("removing_normalizerfull_power.csv")
df_normalizer_off_half_power = pd.read_csv("removing_normalizerhalf_power.csv")
df_normalizer_off_half_power_half_power = pd.read_csv("removing_normalizerhalf_powerhalf.csv")
df_normalizer_off_half_powerquarter = pd.read_csv("removing_normalizerhalf_powerquarter.csv")
df_power_reallocation = pd.read_csv("power_reallocation.csv")
df_normlizeroff1100 = pd.read_csv("normalizer_off_1100.csv")
df_normlizeroff1111 = pd.read_csv("normalizeroff_1111.csv")
```

```python

plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
sns.lineplot(data=df_half_power[df_half_power['Metric'] == 'Throughput'], x='gnbUeDistance', y='Value', label='Half Power', color='blue')
sns.lineplot(data=df_full_power[df_full_power['Metric'] == 'Throughput'], x='gnbUeDistance', y='Value', label='Full Power', color='orange')
sns.lineplot(data=df_normal_run_allegation[df_normal_run_allegation['Metric'] == 'Throughput'], x='gnbUeDistance', y='Value', label='Normal Run Allegation', color='green')
sns.lineplot(data=df_upnormal_run_allegation[df_upnormal_run_allegation['Metric'] == 'Throughput'], x='gnbUeDistance', y='Value', label='full power 8x8', color='red')
sns.lineplot(data=df_upnormal_run_allegation_16half[df_upnormal_run_allegation_16half['Metric'] == 'Throughput'], x='gnbUeDistance', y='Value', label='half power  16x16', color='purple')
sns.lineplot(data=df_upnormal_run_allegation_16[df_upnormal_run_allegation_16['Metric'] == 'Throughput'], x='gnbUeDistance', y='Value', label='full power  16x16', color='black')


plt.title('Distance vs througthput')
plt.xlabel('Distance (m)')
plt.ylabel('throughput (Mbps)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 7))  # Increased figure size for better readability
sns.set(style="whitegrid")

# Plot the data with distinct line styles and colors
sns.lineplot(data=df_half_power[df_half_power['Metric'] == 'Throughput'], 
             x='gnbUeDistance', y='Value', 
             label='Half Power', color='blue', linestyle='solid')
sns.lineplot(data=df_full_power[df_full_power['Metric'] == 'Throughput'], 
             x='gnbUeDistance', y='Value', 
             label='Full Power', color='orange', linestyle='dashed')
sns.lineplot(data=df_normal_run_allegation[df_normal_run_allegation['Metric'] == 'Throughput'], 
             x='gnbUeDistance', y='Value', 
             label='Normal Run Allegation', color='green', linestyle='dashdot')
sns.lineplot(data=df_upnormal_run_allegation[df_upnormal_run_allegation['Metric'] == 'Throughput'], 
             x='gnbUeDistance', y='Value', 
             label='Full Power 8x8 MIMO', color='red', linestyle='solid')
sns.lineplot(data=df_upnormal_run_allegation_16half[df_upnormal_run_allegation_16half['Metric'] == 'Throughput'], 
             x='gnbUeDistance', y='Value', 
             label='Half Power 16x16 MIMO', color='purple', linestyle='dotted')
sns.lineplot(data=df_upnormal_run_allegation_16[df_upnormal_run_allegation_16['Metric'] == 'Throughput'], 
             x='gnbUeDistance', y='Value', 
             label='Full Power 16x16 MIMO', color='black', linestyle='dashed')

# Enhance title and labels
plt.title('Throughput vs. Distance for MIMO Configurations in V2V Highway Scenario\nGenerated: May 22, 2025', 
          pad=20, fontsize=14, weight='bold')
plt.xlabel('Distance (m)', fontsize=12)
plt.ylabel('Throughput (Mbps)', fontsize=12)

# Customize legend
plt.legend(title='Configuration Types', loc='upper right', bbox_to_anchor=(1.15, 1), 
           fontsize=10, title_fontsize=12)

# Add grid and adjust limits
plt.grid(True, linestyle='--', alpha=0.7)
plt.ylim(50, 210)  # Truncate y-axis to focus on relevant range, add note if needed
plt.xticks(range(0, 1100, 100))  # Add more tick marks for granularity

# Add annotations for key points
plt.annotate('Max: 200 Mbps', xy=(0, 200), xytext=(0, 210),
             arrowprops=dict(facecolor='black', shrink=0.05), fontsize=10)
plt.annotate('Drop to 100 Mbps', xy=(500, 100), xytext=(550, 110),
             arrowprops=dict(facecolor='black', shrink=0.05), fontsize=10)

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(12, 8))  # Increased figure size for better readability
sns.set(style="whitegrid")

# Define more distinct colors, considering colorblindness
colors = {
    'Half Power': 'blue',
    'Full Power': 'orange',
    'Normal Run Allegation': 'green',
    'Full Power 8x8': 'red',
    'Half Power 16x16': 'purple',
    'Full Power 16x16': 'black',
}

# Plotting with defined colors and clearer labels
sns.lineplot(
    data=df_half_power[df_half_power['Metric'] == 'Throughput'],
    x='gnbUeDistance',
    y='Value',
    label='Half Power',
    color=colors['Half Power'],
    linewidth=2,
)  # Increased linewidth
sns.lineplot(
    data=df_full_power[df_full_power['Metric'] == 'Throughput'],
    x='gnbUeDistance',
    y='Value',
    label='Full Power',
    color=colors['Full Power'],
    linewidth=2,
)
sns.lineplot(
    data=df_normal_run_allegation[
        df_normal_run_allegation['Metric'] == 'Throughput'
    ],
    x='gnbUeDistance',
    y='Value',
    label='Normal Run Allegation',
    color=colors['Normal Run Allegation'],
    linewidth=2,
)
sns.lineplot(
    data=df_upnormal_run_allegation[
        df_upnormal_run_allegation['Metric'] == 'Throughput'
    ],
    x='gnbUeDistance',
    y='Value',
    label='Full Power 8x8',
    color=colors['Full Power 8x8'],
    linewidth=2,
)
sns.lineplot(
    data=df_upnormal_run_allegation_16half[
        df_upnormal_run_allegation_16half['Metric'] == 'Throughput'
    ],
    x='gnbUeDistance',
    y='Value',
    label='Half Power 16x16',
    color=colors['Half Power 16x16'],
    linewidth=2,
)
sns.lineplot(
    data=df_upnormal_run_allegation_16[
        df_upnormal_run_allegation_16['Metric'] == 'Throughput'
    ],
    x='gnbUeDistance',
    y='Value',
    label='Full Power 16x16',
    color=colors['Full Power 16x16'],
    linewidth=2,
)


plt.title('Throughput vs. Distance for Different Configurations', fontsize=16)  # More descriptive title, larger font size
plt.xlabel('Distance (m)', fontsize=14)  # Larger font size for labels
plt.ylabel('Throughput (Mbps)', fontsize=14)
plt.legend(
    title='Configuration',
    fontsize=12,
    title_fontsize=13,
)  # Legend with title, larger font
plt.xticks(fontsize=12)  # Larger tick label font
plt.yticks(fontsize=12)
plt.grid(True)
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```
