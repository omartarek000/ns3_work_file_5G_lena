```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_norm_on_default = pd.read_csv("norm_on_default.csv") # default normlizer on no port powe configuration 
df_norm_on_port1111 = pd.read_csv("norm_on_port_1111.csv") # default normlizer on port 1111 configuration
df_norm_on_port1100 = pd.read_csv("norm_on_port_1100.csv") # default normlizer on port 1100 configuration
df_norm_on_port050500 =  pd.read_csv("norm_on_port_050500.csv") # default normlizer on port 050500 configuration
df_norm_on_port02502500 = pd.read_csv("norm_on_port_02502500.csv") # default normlizer on port 02502500 configuration
df_norm_on_default_port_1x1 = pd.read_csv("norm_on_default_port(1x1).csv") # default normlizer on default port 1x1 configuration
```

```python
# Set professional styling
plt.style.use('seaborn-v0_8-whitegrid')
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))

# Define consistent color scheme and styling for all configurations
config_styles = {
    'default': {'color': '#1f77b4', 'linestyle': '--', 'linewidth': 2.5, 'label': 'Default Configuration'},
    'port1111': {'color': '#ff7f0e', 'linestyle': '-', 'linewidth': 2.5, 'label': 'Full Power (1,1,1,1)'},
    'port1100': {'color': '#2ca02c', 'linestyle': '-.', 'linewidth': 2.5, 'label': 'Reduced Power (1,1,0,0)'},
    'port050500': {'color': '#d62728', 'linestyle': ':', 'linewidth': 2.5, 'label': 'Low Power (0.5,0.5,0,0)'},
    'port02502500': {'color': '#9467bd', 'linestyle': '-.', 'linewidth': 2.5, 'label': 'Mixed Power (0.25,0.25,0.5,0)'}
}

# LEFT PLOT: Power Consumption Analysis
ax1.set_title('Power Consumption Analysis', fontsize=14, fontweight='bold', pad=20)

# Calculate average power consumption for each configuration
avg_powers = {
    'default': df_norm_on_default['Power_W'].mean() if 'Power_W' in df_norm_on_default.columns else 120,  # placeholder
    'port1111': power_data1111['EffectivePower_W'].mean(),
    'port1100': power_data1100['EffectivePower_W'].mean(),
    'port1000': power_data1000['EffectivePower_W'].mean() if 'power_data1000' in globals() else 40,  # placeholder
}

# Create bar chart for average power consumption
configs = ['Default', 'Full Power\n(1,1,1,1)', 'Reduced Power\n(1,1,0,0)', 'Minimal Power\n(1,0,0,0)']
powers = [avg_powers['default'], avg_powers['port1111'], avg_powers['port1100'], avg_powers['port1000']]
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#17becf']

bars = ax1.bar(configs, powers, color=colors, alpha=0.8, edgecolor='white', linewidth=1.5)

# Add value labels on bars
for bar, power in zip(bars, powers):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 2,
             f'{power:.1f}W', ha='center', va='bottom', fontweight='bold', fontsize=10)

# Add percentage difference annotations
reference_power = powers[1]  # Full power as reference
for i, (bar, power) in enumerate(zip(bars[2:], powers[2:]), start=2):
    pct_diff = ((reference_power - power) / reference_power) * 100
    ax1.annotate('', xy=(1, reference_power), xytext=(i, power),
                arrowprops=dict(arrowstyle='<->', color='red', alpha=0.6, lw=1.5))
    
    # Position text between the bars
    mid_y = (reference_power + power) / 2
    ax1.text((1 + i) / 2, mid_y, f'-{pct_diff:.0f}%', 
             ha='center', va='center', fontweight='bold', fontsize=9,
             bbox=dict(boxstyle='round,pad=0.3', facecolor='white', edgecolor='red', alpha=0.8))

ax1.set_ylabel('Average Power Consumption (W)', fontsize=11, fontweight='semibold')
ax1.set_xlabel('Configuration', fontsize=11, fontweight='semibold')
ax1.grid(True, alpha=0.3, axis='y')
ax1.set_ylim(0, max(powers) * 1.2)

# RIGHT PLOT: Throughput vs Distance
ax2.set_title('Throughput vs Distance Performance', fontsize=14, fontweight='bold', pad=20)



sns.lineplot(data=df_norm_on_port1111[df_norm_on_port1111['Metric'] == 'Throughput'], 
             x='gnbUeDistance', y='Value',
             color=config_styles['port1111']['color'], 
             linestyle=config_styles['port1111']['linestyle'],
             linewidth=config_styles['port1111']['linewidth'],
             label=config_styles['port1111']['label'], ax=ax2)

sns.lineplot(data=df_norm_on_port1100[df_norm_on_port1100['Metric'] == 'Throughput'], 
             x='gnbUeDistance', y='Value',
             color=config_styles['port1100']['color'], 
             linestyle=config_styles['port1100']['linestyle'],
             linewidth=config_styles['port1100']['linewidth'],
             label=config_styles['port1100']['label'], ax=ax2)



ax2.set_xlabel('Distance (m)', fontsize=11, fontweight='semibold')
ax2.set_ylabel('Throughput (Mbps)', fontsize=11, fontweight='semibold')
ax2.legend(fontsize=9, frameon=True, fancybox=True, shadow=True, loc='upper right')
ax2.grid(True, alpha=0.3)

# Overall styling
fig.suptitle('Power Efficiency vs Performance Analysis', fontsize=16, fontweight='bold', y=0.95)

# Add subtle background
for ax in [ax1, ax2]:
    ax.set_facecolor('#f8f9fa')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_color('#cccccc')
    ax.spines['bottom'].set_color('#cccccc')

plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.show()
```
